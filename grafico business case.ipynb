```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dict_baseline = {"suppliers": [1, 2, 3, 4], "Environmental": [0.0373, 0.2945, -0.2743, -0.0575], "Social": [-0.4100, -0.2588, 0.4141, 0.2547], "rank": [1000,1000,1000,1000]}
dict_sc1 = {"suppliers": [1, 2, 3, 4], "Environmental": [0.3918, -0.0802, -0.4870, 0.1754], "Social": [-0.4165, -0.0217, 0.3155, 0.1226], "rank": [1000,1000,1000,1000]}
dict_sc2 = {"suppliers": [1, 2, 3, 4], "Environmental": [0.0165, -0.1754, 0.1325, 0.0264], "Social": [0.3136, -0.4116, 0.3731, -0.2750], "rank":[1000,1000,1000,1000]}

df_baseline = pd.DataFrame(data=dict_baseline)
df_baseline["RankGlobal"] = df_baseline["Environmental"] * 0.449 + df_baseline["Social"] * 0.551

df_sc1 = pd.DataFrame(data=dict_sc1)
df_sc1["RankGlobal"] = df_sc1["Environmental"] * 0.449 + df_sc1["Social"] * 0.551

df_sc2 = pd.DataFrame(data=dict_sc2)
df_sc2["RankGlobal"] = df_sc2["Environmental"] * 0.449 + df_sc2["Social"] * 0.551

df_baseline.head()

sup1_bs = df_baseline.iloc[0:1]
sup2_bs = df_baseline.iloc[1:2]
sup3_bs = df_baseline.iloc[2:3]
sup4_bs = df_baseline.iloc[3:4]

sup1_sc1 = df_sc1.iloc[0:1]
sup2_sc1 = df_sc1.iloc[1:2]
sup3_sc1 = df_sc1.iloc[2:3]
sup4_sc1 = df_sc1.iloc[3:4]

sup1_sc2 = df_sc2.iloc[0:1]
sup2_sc2 = df_sc2.iloc[1:2]
sup3_sc2 = df_sc2.iloc[2:3]
sup4_sc2 = df_sc2.iloc[3:4]
```

```python
# plot
fig, ax = plt.subplots(figsize=(20,10))

ax.scatter(sup1_bs["Environmental"], sup1_bs["Social"], c='b', s=sup1_bs["rank"], edgecolors = "w", label = "Supplier 1")
ax.scatter(sup2_bs["Environmental"], sup2_bs["Social"], c='g', s=sup2_bs["rank"], edgecolors = "w", label = "Supplier 2")
ax.scatter(sup3_bs["Environmental"], sup3_bs["Social"], c="r", s=sup3_bs["rank"], edgecolors = "w", label = "Supplier 3")
ax.scatter(sup4_bs["Environmental"], sup4_bs["Social"], c="m", s=sup4_bs["rank"], edgecolors = "w", label = "Supplier 4")

ax.scatter(sup1_sc1["Environmental"], sup1_sc1["Social"], c='b', s=sup1_sc1["rank"], marker=r'$\alpha_1$', label = "Scenario 1, Sup 1")
ax.scatter(sup2_sc1["Environmental"], sup2_sc1["Social"], c='g', s=sup2_sc1["rank"], marker=r'$\alpha_1$', label = "Scenario 1, Sup 2")
ax.scatter(sup3_sc1["Environmental"], sup3_sc1["Social"], c="r", s=sup3_sc1["rank"], marker=r'$\alpha_1$', label = "Scenario 1, Sup 3")
ax.scatter(sup4_sc1["Environmental"], sup4_sc1["Social"], c="m", s=sup4_sc1["rank"], marker=r'$\alpha_1$', label = "Scenario 1, Sup 4")

ax.scatter(sup1_sc2["Environmental"], sup1_sc2["Social"], c='b', s=sup1_sc2["rank"], marker=r'$\alpha_2$', label = "Scenario 2, Sup 1")
ax.scatter(sup2_sc2["Environmental"], sup2_sc2["Social"], c='g', s=sup2_sc2["rank"], marker=r'$\alpha_2$', label = "Scenario 2, Sup 2")
ax.scatter(sup3_sc2["Environmental"], sup3_sc2["Social"], c="r", s=sup3_sc2["rank"], marker=r'$\alpha_2$', label = "Scenario 2, Sup 3")
ax.scatter(sup4_sc2["Environmental"], sup4_sc2["Social"], c="m", s=sup4_sc2["rank"], marker=r'$\alpha_2$', label = "Scenario 2, Sup 4")

#Criao da legenda
ax.legend(bbox_to_anchor=(1.05, 1),
                         loc='upper left', fontsize=15, markerscale = 0.5)

#ttulos do eixo x com a seta.
ax.text(
    0, -1.2, "                 Environment                  ", ha="center", va="center", rotation=0, size=30, color = "w",
    bbox=dict(boxstyle="rarrow,pad=0.2", fc="green", ec="w", lw=2))

#ttulos do eixo y com a seta.
ax.text(
    -1.2, 0, "                      Social                      ", ha="center", va="center", rotation=90, size=30, 
    bbox=dict(boxstyle="rarrow,pad=0.2", fc="yellow", ec="black", lw=2))

#ttulos do grfico.
ax.set_title("Sustainability category of suppliers", fontsize=20)


#Seta para o supplier 1 - 2022 -> sc1
ax.annotate("",
            xy=(sup1_sc1["Environmental"][0], sup1_sc1["Social"][0]), xycoords='data',
            xytext=(sup1_bs["Environmental"][0], sup1_bs["Social"][0]), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )
#Seta para o supplier 1 - 2022 -> sc2
ax.annotate("",
            xy=(sup1_sc2["Environmental"][0], sup1_sc2["Social"][0]), xycoords='data',
            xytext=(sup1_bs["Environmental"][0], sup1_bs["Social"][0]), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )

#Seta para o supplier 2 - 2022 -> sc1
ax.annotate("",
            xy=(sup2_sc1["Environmental"][1], sup2_sc1["Social"][1]), xycoords='data',
            xytext=(sup2_bs["Environmental"][1], sup2_bs["Social"][1]), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )
#Seta para o supplier 2 - 2022 -> sc2
ax.annotate("",
            xy=(sup2_sc2["Environmental"][1], sup2_sc2["Social"][1]), xycoords='data',
            xytext=(sup2_bs["Environmental"][1], sup2_bs["Social"][1]), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )

#Seta para o supplier 3 - 2022 -> sc1
ax.annotate("",
            xy=(sup3_sc1["Environmental"][2], sup3_sc1["Social"][2]), xycoords='data',
            xytext=(sup3_bs["Environmental"][2], sup3_bs["Social"][2]), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )
#Seta para o supplier 3 - 2022 -> sc2
ax.annotate("",
            xy=(sup3_sc2["Environmental"][2], sup3_sc2["Social"][2]), xycoords='data',
            xytext=(sup3_bs["Environmental"][2], sup3_bs["Social"][2]), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )

#Seta para o supplier 4 - 2022 -> sc1
ax.annotate("",
            xy=(sup4_sc1["Environmental"][3], sup4_sc1["Social"][3]), xycoords='data',
            xytext=(sup4_bs["Environmental"][3], sup4_bs["Social"][3]), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )
#Seta para o supplier 4 - 2022 -> sc2
ax.annotate("",
            xy=(sup4_sc2["Environmental"][3], sup4_sc2["Social"][3]), xycoords='data',
            xytext=(sup4_bs["Environmental"][3], sup4_bs["Social"][3]), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )


img = plt.imread("/Users/felipedutra/Documents/Curso_Mestrado_DEI/2098_MCDM/Business Case/fundo.png")
ax.imshow(img, extent=[-1, 1, -1, 1], alpha=0.4)
plt.savefig("grafico_ranking.png")
#plt.savefig("grafico_ranking_baseline.png")
plt.show()
```
